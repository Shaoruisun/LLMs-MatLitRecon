```python
#单种催化剂

with open('./10.1016j.apcata.2004.01.001..json') as f:
  data = json.load(f)
```

```python
# 创建Digraph对象
dot = graphviz.Digraph(comment='Process Flow', graph_attr={'rankdir': 'LR'})

# 定义绘图函数
def draw_graph(graph, node, parent_name=None, step_count=0, **kwargs):
    for key, value in node.items():
        if isinstance(value, dict):
            # 绘制第二级key为主线的节点
            graph.node(key)
            if parent_name:
                # 将冒号前的信息放在线上
                edge_label = key
                graph.edge(parent_name, key)#label=edge_label)
            # 处理step中的key，按顺序添加有向箭头
            if key == 'step':
                step_keys = list(value.keys())
                for i in range(len(step_keys) - 1):
                    step_count += 1
                    graph.edge(step_keys[i], step_keys[i + 1], dir='forward', constraint='false', label=f'step{step_count}')
            # 递归绘制子图
            draw_graph(graph, value, parent_name=key, step_count=step_count, **kwargs)
        else:
            # 将所有信息表示为节点
            info_node_name = f'{parent_name}_{key}'
            graph.node(info_node_name, f'{value}')
            # 将冒号前的信息放在线上
            edge_label = key
            graph.edge(parent_name, info_node_name, label=edge_label)

# 调用绘图函数
draw_graph(dot, data)

# 渲染和保存图像
output_file = 'process_flow'
dot.render(output_file, format='png', cleanup=True)
```

```python

directory_path = ' your_directory_path'
output_folder = './ProcessedSyntheticRoute'
os.makedirs(output_folder, exist_ok=True)
dot = graphviz.Digraph(comment='Process Flow', graph_attr={'rankdir': 'LR'})
unprocessed_file_len = []
unprocessed_file_error = []
for filename in os.listdir(directory_path):
    print(filename)
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
        if len(data) > 10:
            unprocessed_file_len.append(filename)
            print(filename,'   长度大于12')
            
        else:
            try:
                #dot = graphviz.Digraph(comment='Process Flow', graph_attr={'rankdir': 'LR'})
                for catalyst_name, catalyst_info in data.items():
                    dot = graphviz.Digraph(comment='Process Flow', graph_attr={'rankdir': 'LR'})
                    dicts = {catalyst_name: catalyst_info}
                    draw_graph(dot, dicts)                      
                    output_file = os.path.join(output_folder,filename[:-6]+catalyst_name)
                    dot.render(output_file, format='png', cleanup=True)
            except AttributeError as e:
                unprocessed_file_error.append(filename)
                
                print(f"Error processing file {filename}: {type(e).__name__} - {e}")
                continue
        
        
```
